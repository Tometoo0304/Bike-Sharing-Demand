```python
import requests
import datetime
import os
import selenium
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0"
firefox_driver = os.path.join(os.getcwd(),"Drivers","geckodriver.exe")
firefox_service = Service(firefox_driver)
firefox_options = Options()
firefox_options.add_argument('--disable-blink-features=AutomationControlled')
firefox_options.set_preference('general.useragent.override', user_agent)
browser = webdriver.Firefox(service = firefox_service, options = firefox_options)
weather_2017 = []
start_date_2017 = datetime.date(2017,1,1)
end_date_2017 = datetime.date(2017,12,31)
while start_date_2017 <= end_date_2017:
    url = "https://www.wunderground.com/history/daily/us/va/arlington/KDCA/date/"+str(start_date_2017)
    browser.get(url)
    tables = WebDriverWait(browser,20).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "mat-table.cdk-table.mat-sort.ng-star-inserted")))
    weather = pd.read_html(tables[-1].get_attribute('outerHTML'))[0]
    weather = weather.dropna(how = "all")
    weather['Date'] = start_date_2017
    weather_2017.append(weather)
    start_date_2017 = start_date_2017 + datetime.timedelta(1)
weather_2017 = pd.concat(weather_2017, axis = 0)
```

```python
weather_2017["Date"] = pd.to_datetime(weather_2017['Date'], format='%Y-%m-%d')
weather_2017.to_csv(r"C:\Users\wangj\Desktop\Bike_Dataset\weather_2017.csv", index = False, encoding = 'utf-8')
```
